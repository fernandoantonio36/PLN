## **Sobre a base de dados**
### Digit Recognizer - A famosa base de dados MNIST


MNIST ("Modified National Institute of Standards and Technology"), que existe desde 1999, é a base de dados tradicional utilizada. Trata-se de um conjunto de dados de imagens de dígitos, de zero a nove, escritos à mão. Cada imagem, 28 x 28, em pixels, está em escala de cinza e somam um total de 784 pixels. Cada pixel tem um único valor associado, indicando mais claro ou mais escuro, com números maiores significando mais escuro. Esse valor de pixel é um inteiro entre 0 e 255, inclusive.
O conjunto de dados de treinamento (train.csv) tem 785 colunas. A primeira coluna, chamada "label", é o dígito que foi desenhado pelo usuário. O resto das colunas contêm os valores de pixel da imagem associada. A única diferença entre esse conjunto de dados e o de test (test.csv), é que este último não possui a coluna "label". A base de dados test não será utilizada nesta atividade, pois utilizaremos outra base de teste a partir do train.csv na etapa de divisão da base em treino, validação e teste. 

## **Introdução**

A visão computacional é o campo da Inteligência Artificial que permite que computadores e sistemas obtenham informações significativas de imagens digitais, vídeos e outras entradas visuais e tomem ações ou façam recomendações com base nessas informações. Se a IA permite que os computadores pensem, a visão computacional permite que eles vejam, observem e compreendam. A visão computacional precisa de muitos dados. Ela executa análises de dados repetidamente até compreender as distinções e, finalmente, reconhecer imagens. Por exemplo, para treinar um computador para reconhecer pneus de automóveis, ele precisa ser alimentado com grandes quantidades de imagens de pneus e itens relacionados a pneus para aprender as diferenças e reconhecer um pneu, especialmente um sem defeitos. Duas tecnologias essenciais são usadas para fazer isso:

*   Aprendizado profundo (Deep Learning)
*   Rede neural convolucional (CNN)

Uma rede neural consiste em neurônios que são organizados em camadas. Os neurônios recebem um vetor de entrada e o convertem em uma saída. O processo envolve cada neurônio recebendo entrada e aplicando uma função que geralmente é uma função não linear, e em seguida, passa a saída para a próxima camada, conforme imagem seguinte.

camadas_an.gif

Essa atividade utilizará uma Rede Neural Convolucional (CNN) como técnica para Visão Computacional para classificação da base de dados MNIST. As CNN são modelos inspirados na biologia do córtex visual dos mamíferos. Hubel e Wiesel (1960) propuseram explicação para a forma com que os mamíferos percebem visualmente o mundo ao seu redor utilizando arquitetura em camadas de neurônios no cérebro, e isso, por sua vez, inspirou os engenheiros a desenvolver mecanismos semelhantes de reconhecimento de padrões na visão computacional. A figura abaixo, inspirada por DiCarlo e Cox (2007), ilustra as transformações que ocorrem no fluxo visual ventral.

image.png

Na hipótese dos autores citados, dentro do córtex visual, respostas funcionais complexas geradas por células "profundas" são construídas a partir de respostas simples de "células iniciais". Por exemplo, células simples responderão a linhas especificamente orientadas e bordas, enquanto células profundas também responderão a certas orientações, mas com grau progressivo de complexidade.

Basicamente uma célula profunda responde a soma de entradas de outras células iniciais. A arquitetura das redes neurais convolucionais foi inspirada nestas idéias.

## **Passo 3 - Transformando os dados que estão no formato csv em matrizes**
**O conjunto de dados train.csv será convertido em matrizes.**


É necessário informar explicitamente em qual categoria as imagens pertecem em uma lista denominada labels (dígitos de 0 a 9). Portanto, a coluna labels será convertida num array. Após isso, iremos converter as imagens das bases de treinamento (train.csv) em matrizes 28x28. Isso porque as redes neurais utilizam em suas entradas arrays. Após essa etapa, iremos visualizar 20 imagens do conjunto de imagens de treino escolhidas aleatoriamente. Por fim, faremos a etapa de normalização para evitar variabilidade do dataset. Utilizaremos para essa atividade a biblioteca Numpy abaixo, assim como a Plotly.

```python
import numpy as np
```

```python
#Convertendo a coluna labels num array
labels = np.array(df_train.label) 
```

```python
#Remove a coluna label do dataframe df_train e atualiza esse mesmo dataframe com essa alteração.
df_train.drop(['label'], axis=1, inplace=True) 
```

```python
#Convertendo as imagens numa matriz 28x28
train_array = np.array(df_train).reshape(df_train.shape[0], 28, 28)
```

```python
import matplotlib.pyplot as plt
```

```python
# Visualizing images for labels
plt.figure(figsize=(20,15))
for i in range(20): # Running loop for each digit to be printed
    plt.subplot(5,5,i+1)
    index = np.random.randint(0,42000)
    plt.imshow(train_array[index], cmap = 'Greys')
    plt.title(labels[index])
```

```python
#faremos a normalização, dividindo a média pelo desvio padrão
# normalização diminui a variabilidade do dataset, deixando os valores mais próximos um do outro
for i in range(len(train_array)):
  train_array[i] = (train_array[i] - np.average(train_array[i], axis= (0, 1))) / np.std(train_array[i], axis= (0, 1)) 
```

## **Passo 4 - Divisão treino, teste e validação**

**Será utilizado o método train_test_split para dividir os dados em um conjunto treino, teste e validação.**
Utilizaremos a biblioteca sklearn para essa atividade

```python
#Vamos dividir o grupo de treino, validação e teste na proporção de cerca de 70%/15%/15%
#A quantidade de imagens em cada grupo ficará da seguinte forma, respectivamente:
# 29.400/6.300/6.300
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
dataset_train, dataset_test, labels_train, labels_test = train_test_split(train_array, labels, test_size=0.15, random_state=80)
dataset_train, dataset_val, labels_train, labels_val = train_test_split(dataset_train, labels_train, test_size=0.15, random_state=80)
```

## **Passo 5 - Criando a estrutura da Rede Neural Convolucional**

**Essa etapa cria a arquitetura da Rede Neural Convolucional utilizada neste projeto.**

Utilizaremos vários pacotes da biblioteca keras conforme segue abaixo.

```python
#Nesta etapa vamos criar a arquitetura da nossa rede neural convolucional, 
#Utilizaremos a biblioteca Keras, própria para Deep Learning em Python
#Inicialmente vamos importar as funções do Keras que iremos utilizar:

from keras.models import Sequential
from keras import optimizers
from keras.layers.core import Dense, Dropout
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPool2D, GlobalAveragePooling2D
from keras.models import Model
from keras.layers import Input, Concatenate, add
from keras.callbacks import ModelCheckpoint
from keras.layers import Activation, Dense, LeakyReLU
from keras.utils.np_utils import to_categorical
import keras
from keras.layers import Flatten
```

```python
#Agora criaremos a estrutura da rede neural convolucional
# Definindo o modelo Sequential, ideal para conectar uma lista de camadas
model = Sequential()

# Camadas Conv2D criam um kernel da convolução
model.add(Conv2D(filters=64, kernel_size=(5,5), activation = 'relu', input_shape=(28,28,1)))
model.add(Conv2D(filters = 32, kernel_size=(3,3), activation = 'relu'))
```

```python
# MaxPool2D reduz o tamanho da saída da camada convolucional.
# Isso é feito para evitar que o número de parâmetros da rede aumente demais
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(filters = 32, kernel_size =(3,3), activation = 'relu'))
model.add(Conv2D(filters = 16, kernel_size =(3,3), activation = 'relu'))
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
# Camadas densas recebem a entrada de todos os neurônios da camada anterior. 
#São usadas para classificar a image baseada na saída das camadas convolucionais
model.add(Dense(64, activation = 'relu'))
model.add(Dense(10, activation = 'Softmax', name = 'Output_layer'))
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', # Definindo a função perda
             optimizer = 'adam', # Definindo a função perda
             metrics = ['accuracy']) # Definindo a métrica de avaliação, Acurácia
```

```python
model.summary()
```

## **Passo 6 - Treinando a rede neural**

**Essa etapa irá treinar a rede neural com os dados das imagens.**

Utilizaremos vários pacotes da biblioteca keras conforme segue abaixo e a biblioteca time para medirmos o tempo de treinamento.

```python
import time 
```

```python
print('Treinando a Rede Neural Convolucional:')
start_time = time.time()
hist = model.fit(dataset_train, labels_train, validation_data = (dataset_val, labels_val ), epochs = 20)
#Definimos o treinamento com o dataset de treino, realizando validação no dataset de validação.
#O treinamento não usa o dataset de teste, ficará guardado para avaliarmos nossa rede depois.

tempo = str(time.strftime("%H:%M:%S", time.gmtime(time.time() - start_time)))
print('')
print('Treino finalizado em %s' % tempo)
```

```python
plt.plot(hist.history['loss'], 'b-', label='train loss')
plt.plot(hist.history['val_loss'], 'r-', label='val loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()

#Por fim, plotamos os resultados de evolução da medida de erro (loss) e acurácia ao longo dos epochs
plt.plot(hist.history['loss'], 'b-', label='train loss')
plt.plot(hist.history['val_loss'], 'r-', label='val loss')
plt.ylabel('Perda')
plt.xlabel('epoch')
plt.show()

plt.plot(hist.history['accuracy'], 'b-', label='train accuracy')
plt.plot(hist.history['val_accuracy'], 'r-', label='val accuracy')
plt.ylabel('Acurácia')
plt.xlabel('epoch')
plt.show()

print ('\nAgora vamos avaliar o modelo no dataset de teste. Vamos para a próximo comando.')
```

## **Passo 7 - Avaliando o modelo**
Nessa etapa, o modelo será avaliado a partir do conjunto de testes.

```python
model.evaluate(dataset_test,labels_test)
```

**A acurácia dos dados de teste sobre o modelo está acima de 98%, o que significa que o modelo está funcionando sem overfitting.**
